# Метрика

Перед решением любой задачи важно понимать, как будет оцениваться ваше решение. В данном случае мы используем стандартную для задачи классификации метрику ROC-AUC. Ее можно вычислить, используя только предсказанные вероятности и истинные классы без конкретного порога классификации + она раотает даже если классы в данных сильно несбалансированны (примеров одного класса в десятки раз больше примеров длугого). Именно поэтому она очень удобна для соревнований.

Посчитать ее легко:


```python
from sklearn.metrics import roc_auc_score

y_true = [
    0,
    1,
    1,
    0,
    1
]

y_predictions = [
    0.1,
    0.9,
    0.4,
    0.6,
    0.61
]

roc_auc_score(y_true, y_predictions)
```

```python
import sklearn
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline
from sklearn.compose import ColumnTransformer
```

```python
from sklearn.compose import make_column_transformer
# Определение тренировочного датасета
x_train = data.drop(target_col, axis=1)
y_train = data[target_col]

# # normalization
# pipeline = ColumnTransformer([
#     ('num', StandardScaler, num_cols),
#     ('cat', OneHotEncoder, cat_cols),
# ])
# cls = make_pipeline(
#     pipeline,
#     LogisticRegression
#     )

transformer = make_column_transformer((OneHotEncoder(handle_unknown='ignore'), cat_cols))
transformed = transformer.fit_transform(x_train)
# transformed = transformer.fit_transform(y_train)
# Transformating back
transformed_df = pd.DataFrame(transformed, columns=transformer.get_feature_names_out())
# One-hot encoding removed an index. Let's put it back:
transformed_df.index = x_train.index
# transformed_df.index = y_train.index
# Joining tables
x_train = pd.concat([x_train, transformed_df], axis=1)
# y_train = pd.concat([y_train, transformed_df], axis=1)
# Dropping old categorical columns
x_train.drop(cat_cols, axis=1, inplace=True)
# y_train.drop(cat_cols, axis=1, inplace=True)

# # Transforming
# transformed = transformer.transform(X_test)
# # Transformating back
# transformed_df = pd.DataFrame(transformed, columns=transformer.get_feature_names_out())
# # One-hot encoding removed an index. Let's put it back:
# transformed_df.index = X_test.index
# # Joining tables
# X_test = pd.concat([X_test, transformed_df], axis=1)
# # Dropping old categorical columns
# X_test.drop(categoric_columns, axis=1, inplace=True)

#  Подберём гиперпараметры

clf = LogisticRegression()

params = {
    'C': [100, 10, 1, 0.1, 0.01, 0.001],
}

clf_grid = GridSearchCV(clf, params, cv=5, scoring='roc_auc', refit=True) #verbose - это количество сообщений в выводе
clf_grid.fit(x_train, y_train)

```

```python
from sklearn.metrics import accuracy_score

clf_grid.best_params_

accuracy_score(x_train, clf_grid.best_estimator_.predict(y_train))
```
